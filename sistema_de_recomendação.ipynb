# Pré-Processamento de Dados

## Renomear as colunas

### Tirar informações desnecessárias, como URLs de imagens. Além disso, vamos renomear as colunas de cada arquivo, pois o nome das colunas contem espaços e letras maiusculas. Para facilitar o uso delas, estas correções serão feitas.

```python
##Renomear colunas

books = books[['ISBN', 'Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher']]
books.rename(columns = {'Book-Title': 'title', 'Book-Author': 'author', 'Year-Of-Publication': 'year', 'Publisher': 'publisher'}, inplace=True)
users.rename(columns = {'User-ID': 'user_id', 'Location': 'location', 'Age': 'age'}, inplace=True)
ratings.rename(columns = {'User-ID': 'user_id', 'Book-Rating': 'rating'}, inplace=True)
```

```python
books
```

```python
# Quantidade de rating por usuários
# analise preditiva sera construida com base nas pessoas que estao avaliando
ratings['user_id'].value_counts()
```

```python
#Livros que tenham mais de 200 avaliações
x = ratings['user_id'].value_counts() > 200
x
```

```python
# Quantidade de usuários que fizeram essas avaliações de mais de 200 vezes

y = x[x].index
y.shape
y
```

### Decisão de negócio

```python
# rating de usuarios que avaliaram mais de 200 livros e eliminando todos os outros usuários
ratings = ratings[ratings['user_id'].isin(y)]
```

```python
# juntando tabelas (join ou merge)
# faz um relacionamento entre tabelas por uma chave primária. Neste caso, o 'ISBN'
rating_with_books = ratings.merge(books, on='ISBN')
rating_with_books.head()
```

```python
# quantidade de rating do livro
number_rating = rating_with_books.groupby('title')['rating'].count().reset_index() #sumariza a quantidade de vezes que cada livro foi avaliado
number_rating
```

```python
number_rating.rename(columns = {'rating': 'number_ratings'}, inplace=True)
number_rating
```

```python
from pandas.core.internals.blocks import final
# juntando a tabela de livros com os ratings com a tabela de quantidade de ratings por livro
# varias informações sendo tratadas para chegar numa tabela final
final_rating = rating_with_books.merge(number_rating, on = 'title')
final_rating
```

```python
# filtrar somente os livros que foram avaliados mais de 50 vezes
final_rating = final_rating[final_rating['number_ratings'] >= 50]
final_rating.shape
```

```python
# descartar os valores duplicados, porque se o mesmo usuario tiver avaliado o mesmo livro varias vezes, o modelo preditivo pode dar errado
final_rating.drop_duplicates(['user_id', 'title'], inplace=True)
final_rating.shape
```

```python
final_rating
```

## Transposição de usuários
### Fazer a transposição de usuários em colunas, pois as avaliações dadas por eles serão variáveis para a máquina preditiva
*    Um sistema de recomendação de livros, onde as avaliações foram feitas pelos usuários, então olhar cada avaliação de cada usuário por coluna, já que na linha ficarão os livros avaliados. Cada livro terá nota de cada usuário.

```python
# transposição de linhas(user_id) em colunas
# cada coluna será o usuário, cada linha será o titulo do livro e cada valor será a avaliação(rating) que o usuario fez
book_pivot = final_rating.pivot_table(columns = 'user_id', index = 'title', values = 'rating')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
# tirando os 'NaN', já que a maquina preditiva nao recebe valores nulos
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

### Usaremos o algoritmo de vizinhos mais proximos (nearest neighbors algorithm), que é usada para agrupamento com base na distancia euclidiana.

### Nesta tabela dinamica, muitos valores zero podem aumentar o custo computacional para calcular a distancia dos valores zero, então convertemos a tabela dinamica para uma matriz esparsa.

```python
 from scipy.sparse import csr_matrix
 book_sparse = csr_matrix(book_pivot)
```

# Criação da máquina preditiva

```python
from sklearn.neighbors import NearestNeighbors

# 'model' será a maquina preditiva
model = NearestNeighbors(algorithm='brute')
model.fit(book_sparse)
```

# Novas Predições

```python
book_pivot.head()
```

## Livro 1984

```python
# pegando como exemplo o livro '1984'
distances, suggestions = model.kneighbors(book_pivot.loc['1984', :].values.reshape(1, -1))
```

```python
for i in range(len(suggestions)):
    print(book_pivot.index[suggestions[i]])
```
